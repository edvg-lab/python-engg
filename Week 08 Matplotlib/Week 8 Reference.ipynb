## Pyplot Basics
To add figure, we first need `x` and `y` data

Example:

An object is launched from the ground. Its height from the ground, h(x), after `x` seconds from being launched
is determined by the function:

$h(x) = -4.9x^2 + 16x + 11$


```python
import numpy as np

# numpy array from 0 to 5
x = np.array([0,1,2,3,4,5])
# square of x
y = -4.9*(x**2) + (32*x) + 11
# print values
print(x)
print(y)
```

## Pyplot Basics
To plot the data
- Use the `plt.plot(x,y)` method of matplotlib to add data
- Use `plt.show()` to render the plot

```python
plt.plot(x,y)
plt.show()
```

## Pyplot Basics: Title
- We can add a title to the plot by using the `plt.title(title)` method

```python
plt.plot(x,y)
plt.title('Graph of h(x) = -4.9x^2 + 16x + 11')
plt.show()
```

## Pyplot Basics: Labels
An un-labeled graph is usually unhelpful. We can set labels using the `plt.xlabel(label)` and `plt.ylabel(label)` methods

```python
plt.plot(x,y)
plt.title('Graph of h(x) = -4.9x^2 + 16x + 11')
plt.xlabel('x')
plt.ylabel('h(x)') 
plt.show()
```

## Pyplot Basics: Style
- We can set the line style by adding a third string argument to `plt.plot(x,y,style)`
    - `'-'` solid line
    - `'--'` dashed line
    - `'o'` circle marker
    - more styles in https://matplotlib.org/2.1.2/api/_as_gen/matplotlib.pyplot.plot.html

```python
plt.plot(x,y,'^')
plt.title('Graph of h(x) = -4.9x^2 + 16x + 11')
plt.xlabel('x')
plt.ylabel('h(x)') 
plt.show()
```

## Pyplot Basics: Style
- We can set the line color by prepending a character to the `style` argument `plt.plot(x,y,style)`
    - `'r'` red
    - `'g'` green
    - `'b'` blue
    - more colors in https://matplotlib.org/2.1.2/api/_as_gen/matplotlib.pyplot.plot.html

```python
plt.plot(x,y,'g')
plt.title('Graph of h(x) = -4.9x^2 + 16x + 11')
plt.xlabel('x')
plt.ylabel('h(x)') 
plt.show()
```

## Pyplot Basics: Custom Ticks
We can set the `x` and `y` ticks of the plot by using the `plt.xticks(array-like)` and `plt.xticks(array-like)` methods respectively

```python
plt.plot(x,y,'r--')
plt.title('Graph of h(x) = -4.9x^2 + 16x + 11')
plt.xlabel('x')
plt.ylabel('h(x)') 
plt.show()
```

## Pyplot Basics: Multiple Plots
To add other graphs, simply add more `plt.plot()` before the `plt.show()` method

```python
# x = x^2
plt.plot(x,y,'r--')
# x = x^2
plt.plot(x,x**3)
# x = x
plt.plot(x,x)
plt.title('Multiple Graphs')
plt.xlabel('x')
plt.xticks(x)
plt.show()
```

## Pyplot Basics: Legends
- To make the graph more readable, we can add legends by using the `plt.legend(array-like)` method

```python
plt.plot(x,y,'r--')
plt.plot(x,x**3)
plt.plot(x,x)
plt.title('Multiple Graphs')
plt.xlabel('x')
plt.legend(['-4.9x^2 + 16x + 11','x^3','x'])
plt.show()
```

## Pyplot Basics: Subplots

Sometimes we need to render many graphs in a single feature. For creating subplots, we can use the `plt.subplot(nrows,ncols,index)` method
- index starts at 1

## Sample 2x1 subplots

```python
x = np.arange(0,10,0.1)
x2 = np.arange(0,100,0.1)
y = np.sin(x)
z = np.cos(x2)
plt.subplot(2,1,1)
plt.title("sin(x)")
plt.plot(x,y,'r')

plt.subplot(2,1,2)
plt.title("cos(x)")
plt.plot(x2,z,'g')
plt.xlabel('x')
plt.show()
```

## Sample 2x2 subplots

```python
x = np.arange(0,10,0.1)
y = x**2
z = x**3
z2 = np.exp(x)
```

## Sample 2x2 subplots

```python
plt.subplot(2,2,1) # 0,0
plt.title("x")
plt.plot(x,x,'r')
plt.plot(x,y,'g')

plt.subplot(2,2,2) # 0,1
plt.title("x^2")
plt.plot(x,y,'g')

plt.subplot(2,2,3) # 1,0
plt.title("x^3")
plt.plot(x,z,'b')

plt.subplot(2,2,4) # 1,1
plt.title("exp(x)")
plt.plot(x,z2,'k')

# adjust subplots spacing
plt.subplots_adjust(hspace = 0.4,wspace = 0.4)
plt.show()
```

## Pyplot types: Scatter

```python
x = np.random.random(20)
y = np.random.random(20)
plt.scatter(x,y)
plt.show()
```

## Pyplot types: Pie
`plt.pie(size, explode, labels, color, autopct)` (there are other optional arguments aside from these)

* size = wedge sizes
* explode = optional; specifies the fraction of the radius with which to offset each wedge
* labels = optional strings that labels each wedge
* colors = an optional sequence of colors through which the pie chart will cycle
* autopct = an optional string or function used to label the wedges with their numeric value

```python

label = 'nitrogen', 'argon', 'oxygen', 'trace gases'
sizes = [78, 0.9, 21, 0.1]
explode = (0, 0.5, 0, 0) #only "explodes" the first slice

plt.pie(sizes, explode=explode, labels=label, autopct = '%1.1f%%')

plt.show()
```

## Pyplot types: Bar
