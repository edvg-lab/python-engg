# Using Euler's Method to Approximate Newton's Cooling Law

Newton's cooling law describes how an object of initial temperature $T(t_0) = T_0$ cools down in an environment of temperature $T_R$:<br><br>

$\frac{dT(t)}{dt} = -k (T(t) - T_R)$ <br><br>

It states that the cooling rate $\frac{dT(t)}{dt}$ of the object is proportional to the current temperature difference $(T(t) - T_R)$ to the surrounding environment. Implement Euler's method (with a step size of 10 seconds) to approximate Newton's cooling law.<br><br>

<b>Initial Values:</b><br>
$T_0 = 100$&deg;C <br>
$T_R = 20$&deg;C <br>
$k = 0.07$ <br>
$time\:interval = 0s\: to\: 100s$

```python
# Code here :)

t_0 = 100
t_r = 20
k = 0.07
h = 10

for x in range(0,110,h):
    if x == 0:
        t_1 = t_0
    else:
        d_t = -k*(t_1 - t_r)
        t_1 = t_1 + h*(d_t )
    print("seconds:"+str(x),"temperature:"+str(round(t_1,3)),sep="\t")
```
